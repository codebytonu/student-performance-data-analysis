```python
plt.hist(df["G3"])
plt.title("Final Grade Distribution")
plt.xlabel("G3")
plt.ylabel("Number of Students")
plt.show()
```

```python
df.groupby("studytime")["G3"].mean().plot(kind = "bar")
plt.title("Study Time vs Final Grade")
plt.xlabel("Study time level")
plt.ylabel("Average G3")
plt.show()
```

```python
plt.scatter(df["absences"], df["G3"])
plt.xlabel("Absences")
plt.ylabel("Final Grade(G3)")
plt.title("Absences vs Final Grade")
plt.show()
```

```python
df.groupby("Medu")["G3"].mean().plot(kind = "bar")
plt.title("Mother Education vs Final Grade")
plt.xlabel("Mother Education")
plt.ylabel("Final Grade(G3)")
plt.show()
```

```python
df.groupby("Walc")["G3"].mean().plot(kind = "bar")
plt.title("Weekend Alcohol vs Grade")
plt.xlabel("Weekend Alcohol")
plt.ylabel("Grade")
plt.show()
```

```python
df.groupby("failures")["G3"].mean().plot(kind = "bar")
plt.title("Past Failures vs Grade")
plt.xlabel("Past Failures")
plt.ylabel("Average Grade")
plt.show()
```

```python
df["progress"] = df["G3"] - df["G2"]
df["progress"].head()
plt.hist(df["progress"])
plt.title("Grade Progress(G3 - G2)")
plt.show()
```

```python
corr = df [["G1","G2", "G3", "studytime", "absences", "failures", "Walc", "Dalc"]].corr()
corr
```

```python
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
df["risk"] = (
    (df["failures"] > 0)|
     (df["absences"] > 10)|
      (df["studytime"] == 1)
     )
df["risk"].value_counts()
```

```python
df.groupby("risk")["G3"].mean()
```
